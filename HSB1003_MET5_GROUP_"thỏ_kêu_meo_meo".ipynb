## Chart 4

The question "Does part-time work affect academic performance?" will be answered through our group's main chart. This chart illustrates the correlation between the number of hours students work part-time each week and their self-assessed academic performance in class. As a result, the chart shows a downward trend. The R value is 0.206, indicating a weak relationship between the amount of part-time work and academic performance, with only about 20.6% of the variation in academic performance explained by the time spent on part-time work. This suggests that the correlation between these two factors is weak. Additionally, the density of points shows that most students work part-time between 0 and 20 hours per week, with academic performance ranging from 40 to 80. This indicates high concentration during a moderate amount of working hours. However, at higher working hours (20-30 hours), the point density becomes sparser, showing that fewer students work such long hours. Moreover, the points are scattered along the vertical axis, reflecting that academic performance is not solely influenced by the amount of work, but also depends on many other factors.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
url= 'https://docs.google.com/spreadsheets/d/1J4qpKUwDelcH6fcvcqRjkiZdG95Fm0nGccKaFqF1InQ/export?format=xlsx'
df = pd.read_excel(url, sheet_name='Chart4')
xsales=df['HOURS']
ysales=df['PERCENTAGE']
fit = np.polyfit(xsales, ysales, deg=1)
sns.regplot(x=xsales, y=ysales, color='pink', marker='*',
            scatter_kws={'s': 100, 'alpha': 0.7},
            line_kws={'color': 'blue', 'linestyle': '--'})
plt.text(17, 95, 'y = {0:.3f}x + {1:.3f}'.format(fit[0], fit[1]), fontsize=14, color='red')
plt.text(17, 87, '$R^2$ = {0:.3f}'.format(np.corrcoef(xsales, ysales)[0, 1]**2), fontsize=14, color='red')
plt.title('CORELATION BETWEEN TIME SPENDING FOR WORKING WITH STUDENTS PERFORMANCES IN CLASS', fontsize=9.5)
plt.show()
```

## Chart 5



Through The GPA data and your academic performance, we can clearly see the relationship between GPA and study results. Lower GPA values, such as 1.3, 1.5 and 2.0, are often associated with low performance, with most people scoring only 10 or 20. However, as your GPA increases, academic performance also improves, with more results falling between 40 and 100. Moreover, when the GPA reaches 3.2 or above tend to achieve the highest scores, reflecting excellent academic abilities. The regression line Y shows that a 1-unit increase in GPA raises academic performance evaluation by 38.639 units on average. The R value of 0.681 indicates the model explains 68.1% of the variation, suggesting a moderately strong positive correlation. While the data shows an upward trend, other factors also affect academic performance. To conclude, the higher your GPA, the better your academic performance, showing clear progress and effort in studies.


```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
url= 'https://docs.google.com/spreadsheets/d/1J4qpKUwDelcH6fcvcqRjkiZdG95Fm0nGccKaFqF1InQ/export?format=xlsx'
df = pd.read_excel(url, sheet_name='Chart5')
xsales=df['POINT']
ysales=df['PERCENTAGE']
fit = np.polyfit(xsales, ysales, deg=1)
sns.regplot(x=xsales, y=ysales, color='pink', marker='*',
            scatter_kws={'s': 100, 'alpha': 0.7},
            line_kws={'color': 'blue', 'linestyle': '--'})
plt.text(2.5, 10, 'y={0:.3f} x + {1:.3f}'.format(fit[0], fit[1]), fontsize=14, color='red')
plt.text(2.5, 0, '$R^2$={0:.3f}'.format(np.corrcoef(xsales, ysales)[0, 1]**2), fontsize=14, color='red')
plt.title('CORRELATION BETWEEN GPA AND ACADEMIC PERFORMANCE', fontsize=12)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
url = 'https://docs.google.com/spreadsheets/d/1J4qpKUwDelcH6fcvcqRjkiZdG95Fm0nGccKaFqF1InQ/export?format=xlsx'
df_covid = pd.read_excel(url, sheet_name='Chart 6.1')
x = df_covid['Your GPA (Ever/Never Worked)']
yGP_not_working = df_covid['Your GPA (Ever/Never Worked)']
yGP_working = df_covid['Your GPA (Working)']
plt.plot(x, yGP_not_working, '-r', label='Ever/Never Worked')
plt.plot(x, yGP_working, ':k', label='Working')
plt.xticks(rotation=30, ha='right')
plt.xlabel('GPA OF WORKING PERSON')
plt.ylabel('GPA OF PEOPLE WHO HAVE/ GAVE NEVER WORKED')
plt.title('COMPARE GPA BETWEEN PEOPLE WHO ARE WORKING AND HAVE/ HAVE NOT WORKED')
plt.legend()
plt.tight_layout()
plt.show()
```
